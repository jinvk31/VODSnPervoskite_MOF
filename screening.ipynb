## Import libraries

```python
# import libraries 
import os
import pandas as pd
import pandas
```

## create path for data mining

```python
path = './data/screening_outputs'
if not os.path.exists(path):
    os.makedirs(path)
```

## Import and analyze qmof database  
QMOF database source: https://github.com/Andrew-S-Rosen/QMOF

```python
# read in qmof database and change display settings of pandas dataframes
pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', 400)
qmof_db = pd.read_csv('./data/qmof/qmof.csv',low_memory=False)
qmof_db
```

## Select properties in interest

```python
# for k in qmof_db.columns:
#     print(k)

cols = ['name','info.formula', 'info.doi','info.pld','info.density','info.volume','info.synthesized','info.symmetry.spacegroup','info.symmetry.spacegroup_number','info.symmetry.spacegroup_crystal','info.symmetry.pointgroup','outputs.pbe.bandgap','qmof_id']

qmof = qmof_db[cols]
qmof
```

## Screening by 4 criteria

### criterion 1: pore size

```python
qmof_filtered1 = qmof[qmof['info.pld'].apply(lambda x: 1.7<x<2.8)]
qmof_filtered1.to_csv(os.path.join(path,'1_poresize.csv'))
print(f'{len(qmof_filtered1)} MOFs left')
```

### criterion 2: bandgap energy

```python
qmof_filtered2 = qmof_filtered1[qmof_filtered1['outputs.pbe.bandgap'].apply(lambda x: x>3.0)]
qmof_filtered2.to_csv(os.path.join(path,'2_insulate.csv'))
print(f'{len(qmof_filtered2)} out of {len(qmof_filtered1)} MOFs are insluating materials')
```

### criterion 3: synthesized experimentally 

```python
qmof_filtered3 = qmof_filtered2[qmof_filtered2['info.synthesized']]
qmof_filtered3.to_csv(os.path.join(path,'3_synthesized.csv'))
print(f'{len(qmof_filtered3)} out of {len(qmof_filtered2)} MOFs are insluating materials')
```

### criterion 4: doi link

```python
qmof_filtered4 = qmof_filtered3.dropna()
qmof_filtered4.to_csv(os.path.join(path,'4_doi.csv'))
print(f'{len(qmof_filtered4)} out of {len(qmof_filtered3)} MOFs have doi links')
```

### Filtering out toxic Candidates

```python
import copy
toxic_list = ['As','Pb','Co', 'Cd']
qmof_nontoxic = copy.deepcopy(qmof_filtered4)
```

```python
# check potential toxicity of MOFs
for element in toxic_list:
    num = sum(qmof_filtered4['info.formula'].apply(lambda x: element in x))
    print(f'{num} MOFs out of {len(qmof_filtered4)} is potentially toxic due to {element}')
```

```python
# save non-toxic data
for element in toxic_list:
    qmof_nontoxic = qmof_nontoxic[qmof_nontoxic['info.formula'].apply(lambda x: element not in x)]
    qmof_nontoxic.to_csv(os.path.join(path, f'filtered_{element}.csv'))
    print(f'{len(qmof_nontoxic)} out of {len(qmof_filtered4)} MOFs does not contain {element}')
    
```
