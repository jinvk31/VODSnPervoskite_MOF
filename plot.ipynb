```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from matplotlib.ticker import MultipleLocator
```

```python
mn2_doping_ratio = [0, 1.5, 3, 5]
hall_mobility = [79, 38, 23, 5]
electron_concentration = [9e15, 7e15, 8e14, 5*1e14]
hall_mobility_target = [79, 52, 46, 29.2911]
fig, ax1 = plt.subplots(figsize=(6,6))

x_label = r'$\mathrm{Mn^{2+}}$ doping ratio (mol%)'

ax1.set_xlabel(x_label, fontsize=18);

ax1.set_xlim(-0.5,6 )
y_label1 = r'$\mathrm{Hall\ mobility\ (cm^2V^{-1}s^{-1})}$'
y_tick1 = np.arange(0,130,20)
ax1.set_ylabel(y_label1, fontsize=18);
ax1.set_yticks(y_tick1);
ax1.set_ylim(0,133)
ax1.minorticks_on()

ax1.tick_params(axis='y')
ax1.yaxis.set_minor_locator(MultipleLocator(10))

ax1.tick_params(axis='y',which='major',length = 5, labelsize=16)
ax1.tick_params(axis='y',which='minor',length = 3.5)

ax1.set_box_aspect(1)
bar_width = 0.6
hatch_pattern = '////'

# bar_colors = ['blue', 'orange','red','green']
bar_colors = ['#4A59BE', '#E9892A', '#D1534E', '#439959']
ax1.bar(mn2_doping_ratio, hall_mobility, color=bar_colors, width=bar_width, zorder=2, edgecolor = 'k', alpha=1,hatch=hatch_pattern)

ax2 = ax1.twinx()
for i in range(4):
    ax2.plot(mn2_doping_ratio[i], electron_concentration[i], 'o', markersize=17,color=bar_colors[i], zorder=3,markeredgecolor='k')
y_label2 = r'$\mathrm{Electron\ concentration\ (cm^{-3})}$'
ax2.set_ylabel(y_label2, fontsize=18)
ax2.set_yscale('log')
ax2.set_ylim(1e11, 1e17/1.4);
ax2.tick_params(axis='y',which='major',length = 4, labelsize=16)
ax2.plot(mn2_doping_ratio, electron_concentration, '--', markersize=3, color ='#838383' )

ax1.set_xticks(mn2_doping_ratio);
ax1.tick_params(axis='x',which='major',labelsize=16)
ax1.text(5.8, 128, f'Reference Value', fontsize=12.5, color='black',horizontalalignment='right',verticalalignment='top');
```
